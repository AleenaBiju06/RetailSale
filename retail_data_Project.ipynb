#    Finding the most commonly purchased brands.

```python
top_5_brands=df['Product_Brand'].value_counts().head(5)
print("Top 5 most purchased 5 brands are:")
print(top_5_brands)


plt.figure(figsize=(6,6))
sns.barplot(x=top_5_brands.index,y=top_5_brands.values,palette='coolwarm')
legend_labels=['Pepsi','Adidas','Sony','Samsung','Home Depot']
handles=[plt.Line2D([0],[0],color=color,lw=4)for color in sns.color_palette('coolwarm',5)]
plt.legend(handles,legend_labels,title='Brands')
plt.xticks(rotation=0)
plt.title('Most Commonly Purchased Brands')
plt.xlabel('Product Brand')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

It can be clearly seen that the top 5 most commonly purchased product brands are Pepsi,Adidas,Sony,Samsung and Home Depot.

#   Identifying what type of products are most purchased.

```python

top_product_types = df['Product_Type'].value_counts().head(5)
print("The top 5 most purchased product types are:")
print(top_product_types)

plt.figure(figsize=(6, 6))
sns.barplot(x=top_product_types.values, y=top_product_types.index, palette='Greens')
plt.title('Top 5 Most Commonly Purchased Product Types')
plt.xlabel('count')
plt.ylabel('Product Type')
legend_labels = ['Smartphone', 'Non-Fiction','Fiction','Soft Drinks','Shoes']  
handles=[plt.Line2D([0],[0],color=color,lw=4)for color in sns.color_palette('Greens',5)]
plt.legend(handles,legend_labels,title='Product Types')
plt.show()
```

It shows that the most commonly purchased 5 product types are Smartphone,Fiction,Soft Drink and Shoes.

#   Identifying in which income category most of the customer belongs.

```python
category_counts=df['Income'].value_counts()
print("The income categories are:")
print(category_counts)

sns.set_theme()
colors=sns.color_palette('pastel')
plt.figure(figsize=(10,8))
plt.pie(category_counts,labels=['High','Medium','Low'],autopct='%1.1f%%', startangle=140)
plt.title('Income category Distribution')
plt.axis('equal')
legend_labels = ['High','Medium','Low']  
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in sns.color_palette('pastel', 3)]
plt.legend(handles, legend_labels, title='Income category')
plt.show()
           
```

From the above above graphical representation, we can understand that the most customers are belongs to hingh income category.And the less is in low category.

#    Identifying the purchasing behaviour of different income categories.

```python

purchasing_behavior =df.groupby('Income')['Total_Purchases'].mean()
print("\nAverage purchases by income category:")
print(purchasing_behavior)

plt.figure(figsize=(10,6))
sns.lineplot(x=purchasing_behavior.index, y=purchasing_behavior.values)
plt.title('Average Purchases by Income Category')
plt.xlabel('Income Category')
plt.ylabel('Average Total Purchases')
plt.show()
```

It reveals that the customers of high income category is more likely to purchase products.

#    Detecting in which year the ultimate purchase and the less purchase happened.

```python

yearly_purchases = df.groupby('Year')['Total_Purchases'].sum()
ultimate_purchase_year = yearly_purchases.idxmax()
least_purchase_year = yearly_purchases.idxmin()
print(f"The year with the ultimate purchase: {ultimate_purchase_year} with {yearly_purchases[ultimate_purchase_year]} total purchases.")
print(f"The year with the least purchase: {least_purchase_year} with {yearly_purchases[least_purchase_year]} total purchases.")


plt.figure(figsize=(8, 6))
sns.barplot(x=yearly_purchases.index, y=yearly_purchases.values, palette="Greens")
plt.title('Total Purchases by Year')
plt.xlabel('Year')
plt.ylabel('cont')
legend_labels=['2023','2024']
handles=[plt.Line2D([0],[0],color=color,lw=4)for color in sns.color_palette('Greens',2)]
plt.legend(handles,legend_labels,title='Year')
plt.show()
```

The data reveals that the highest purchase is happened in the year 2023 with 8722.0 total purchases and the least purchase is happened in the year 2024 with 1746.0 total purchases.

#   Analysing  the feedback provided by customers.

```python

feedback_counts = df['Feedback'].value_counts()
print(feedback_counts)


plt.figure(figsize=(8, 5))
feedback_counts.plot(kind='bar', color=['green', 'blue', 'yellow', 'red'])
plt.title('Customer Feedback Distribution')
plt.xlabel('Feedback Type')
plt.ylabel('Number of Responses')
plt.xticks(rotation=0)
plt.show()
```

From the above analysis,we can assume that there is no much difference on the feedbacks provided.There is a mixed satisfaction level is shown in the analysis.

#    Detecting the most applicable shipping method.

```python
Shipping_methods=df['Shipping_Method'].value_counts()
print("The shipping methods are:")
print(Shipping_methods)

plt.figure(figsize=(8, 5))
Shipping_methods.plot(kind='line', color=['skyblue', 'yellow','grey'])
plt.title('Shipping methods')
plt.xlabel('Shipping method')
plt.ylabel('Count')
sns.set_style("dark")
plt.xticks(rotation=0)
plt.show()
```

The data shows the shipping methods adopted by customers for their purchases.It shows that more customers are applied shipping on same-day.

#    Identifyng which payment method is most utilized .

```python

payment_method_counts = df['Payment_Method'].value_counts()
print(payment_method_counts)
most_preferred_payment_method = payment_method_counts.idxmax()
print('Most used method=',most_preferred_payment_method)

sns.set_theme()
colors=sns.color_palette('pastel')
plt.figure(figsize=(8,6))
plt.pie(payment_method_counts,labels=['Debit Card','Credit Card','PayPal','Cash'],autopct='%1.1f%%', startangle=140)
plt.title('Payment Methods Used')
plt.axis('equal')
legend_labels = ['Debit Card','Credit Card','PayPal','Cash']  
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in sns.color_palette('pastel', 4)]
plt.legend(handles, legend_labels, title='Payment Methods')
plt.show()
```

We can see that there is only a minute difference in the payments used by customers.All are almost equally utilized all payment methods.

#   Identifyng in which segmentation  most of the customers belongs.

```python
segment_counts=df['Customer_Segment'].value_counts()
print("Segments","Counts")
print(segment_counts)

plt.figure(figsize=(8, 5))
sns.barplot(x=segment_counts.index, y=segment_counts.values,palette='Reds')

plt.title('Customer Segmentation Counts')
plt.xlabel('Customer Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
legend_labels = ['Regular', 'Premium','New']  
handles = [plt.Line2D([0], [0], color=color, lw=4) for color in sns.color_palette('Reds',3)]
plt.legend(handles, legend_labels, title='Customer Segment')
plt.show()
```

Here,majority customers are belonging to regular category and it shows the customers are regular customers.Only feww number of custoemrs are included in premium and new customer segmentation.

#    Detecting the more amount spend by customer in which category of product.

```python
category_spending = df.groupby('Product_Category')['Amount'].sum().reset_index()
print("Total Spending by Product Category:")
print(category_spending)

plt.figure(figsize=(8, 5))
plt.bar(category_spending['Product_Category'], category_spending['Amount'], color='lightcoral')
plt.title('Spending habit of customers on various Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Purchase Amount')
plt.show()
```

The data reveals the most purchased category of products is from electronics and the least purchase is from home decor.

#   Analysing how the spending habit of male and female is differentiated.

```python
spending_habit= df.groupby(['Gender', 'Product_Category'])['Amount'].sum().unstack(fill_value=0)
print("Total Spending by Gender and Product Category:")
print(spending_habit)

spending_habit.plot(kind='bar', figsize=(10, 6), color=['red','green','blue','yellow','orange'])
plt.title('Total Spending by Gender on various Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Purchase Amount')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

It shows the difference in the spending habit of male and female on various products.Here,male are more likeley to spend for electronics and less in spending for books.Female are more likely to spend almost same for grocery and books.

#   Determining the age category of each customers.

```python
          
age_category = df['Age'].value_counts().sort_index()
print("Customer Distribution by Age Category:")
print(age_category)

plt.figure(figsize=(8, 5))
age_category.plot(kind='line', color='lightgreen')
plt.title('Customer Distribution by Age Category')
plt.xlabel('Age Category')
plt.ylabel('Number of Customers')
plt.show()
```

It can be clearly identified that majority of customers are belongs to the age of 21.
